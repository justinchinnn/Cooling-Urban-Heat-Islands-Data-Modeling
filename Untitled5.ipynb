```python
# -------------------------------
# 3. Mapping Functions for Sentinel-2 and Landsat LST Data
# -------------------------------
def map_satellite_data(tiff_path, csv_path):
    """
    Extracts Sentinel-2 bands (Blue, Green, Red, NIR) based on CSV coordinates,
    and computes NDVI and Albedo.
    """
    data = rxr.open_rasterio(tiff_path)
    tiff_crs = data.rio.crs
    df = pd.read_csv(csv_path)
    latitudes = df['Latitude'].values
    longitudes = df['Longitude'].values
    proj_wgs84 = Proj(init='epsg:4326')
    proj_tiff = Proj(tiff_crs)
    Transformer.from_proj(proj_wgs84, proj_tiff)  # Not used directly
    
    blue_vals, green_vals, red_vals, nir_vals = [], [], [], []
    for lat, lon in tqdm(zip(latitudes, longitudes), total=len(latitudes), desc="Mapping S2 values"):
        blue_vals.append(data.sel(x=lon, y=lat, band=1, method="nearest").values)
        green_vals.append(data.sel(x=lon, y=lat, band=2, method="nearest").values)
        red_vals.append(data.sel(x=lon, y=lat, band=3, method="nearest").values)
        nir_vals.append(data.sel(x=lon, y=lat, band=4, method="nearest").values)
    
    df_sat = pd.DataFrame({
        'Blue': blue_vals,
        'Green': green_vals,
        'Red': red_vals,
        'NIR': nir_vals
    })
    # Compute NDVI = (NIR - Red) / (NIR + Red)
    df_sat['NDVI'] = (df_sat['NIR'] - df_sat['Red']) / (df_sat['NIR'] + df_sat['Red'])
    df_sat['NDVI'] = df_sat['NDVI'].replace([np.inf, -np.inf], np.nan)
    # Compute Albedo as the average of the four bands
    df_sat['Albedo'] = (df_sat['Blue'] + df_sat['Green'] + df_sat['Red'] + df_sat['NIR']) / 4.0
    return df_sat

def map_lst_data(tiff_path, csv_path):
    """
    Extracts Landsat LST values based on CSV coordinates.
    """
    data = rxr.open_rasterio(tiff_path)
    tiff_crs = data.rio.crs
    df = pd.read_csv(csv_path)
    latitudes = df['Latitude'].values
    longitudes = df['Longitude'].values
    proj_wgs84 = Proj(init='epsg:4326')
    proj_tiff = Proj(tiff_crs)
    Transformer.from_proj(proj_wgs84, proj_tiff)
    LST_values = []
    for lat, lon in tqdm(zip(latitudes, longitudes), total=len(latitudes), desc="Mapping LST values"):
        LST_values.append(data.sel(x=lon, y=lat, band=1, method="nearest").values)
    df_lst = pd.DataFrame({'LST': LST_values})
    return df_lst
```

```python
# -------------------------------
# 4. Mapping and Combining Training Data
# -------------------------------
sat_data = map_satellite_data('S2_sample.tiff', '/Users/justin/Downloads/Training_data_uhi_index_2025-02-18.csv')
lst_data = map_lst_data('landsat_lst.tiff', '/Users/justin/Downloads/Training_data_uhi_index_2025-02-18.csv')
def combine_two_datasets(dataset1, dataset2):
    return pd.concat([dataset1, dataset2], axis=1)
train_data = combine_two_datasets(ground_df, sat_data)
train_data = combine_two_datasets(train_data, lst_data)
```

```python
# -------------------------------
# 5. Merge and Aggregate Weather Data for Training
# -------------------------------
# Bronx weather data from CSV.
bronx_weather_df = pd.read_csv("NY_Mesonet_Weather.csv")
print("Bronx weather data sample:")
print(bronx_weather_df.head())
bronx_summary = bronx_weather_df.agg({
    "Air Temp at Surface [degC]": "mean",
    "Avg Wind Speed [m/s]": "mean",
    "Solar Flux [W/m^2]": "mean"
}).to_dict()
train_data["Bronx_AirTemp"]   = bronx_summary["Air Temp at Surface [degC]"]
train_data["Bronx_WindSpeed"] = bronx_summary["Avg Wind Speed [m/s]"]
train_data["Bronx_SolarFlux"] = bronx_summary["Solar Flux [W/m^2]"]

# Manhattan weather data from Excel (sheet "Manhattan").
manhattan_weather_df = pd.read_excel("NY_Mesonet_Weather.xlsx", sheet_name="Manhattan")
print("Manhattan weather data sample:")
print(manhattan_weather_df.head())
manhattan_summary = manhattan_weather_df.agg({
    "Air Temp at Surface [degC]": "mean",
    "Avg Wind Speed [m/s]": "mean",
    "Solar Flux [W/m^2]": "mean"
}).to_dict()
train_data["Manhattan_AirTemp"]   = manhattan_summary["Air Temp at Surface [degC]"]
train_data["Manhattan_WindSpeed"] = manhattan_summary["Avg Wind Speed [m/s]"]
train_data["Manhattan_SolarFlux"] = manhattan_summary["Solar Flux [W/m^2]"]
```

```python
# -------------------------------
# 6. Compute Building Footprint and Urban Morphology Metrics
# -------------------------------
# Load building footprint KML using GeoPandas.
footprint_gdf = gpd.read_file("Building_Footprint.kml", driver="KML")
footprint_gdf = footprint_gdf.to_crs(epsg=32618)

# Create a GeoDataFrame from ground truth coordinates.
gdf_points = gpd.GeoDataFrame(ground_df.copy(), 
                              geometry=gpd.points_from_xy(ground_df.Longitude, ground_df.Latitude),
                              crs="EPSG:4326")
gdf_points = gdf_points.to_crs(epsg=32618)

# Define a buffer around each point (500 meters).
buffer_radius = 500
gdf_points['buffer'] = gdf_points.geometry.buffer(buffer_radius)

# Define functions to compute building metrics.
def compute_building_area(buffer_geom, footprints):
    intersections = footprints.intersection(buffer_geom)
    return intersections.area.sum()

def compute_building_count(buffer_geom, footprints):
    return footprints[footprints.intersects(buffer_geom)].shape[0]

# Compute building metrics for each point.
gdf_points['Building_Area'] = gdf_points['buffer'].apply(lambda b: compute_building_area(b, footprint_gdf))
gdf_points['Building_Count'] = gdf_points['buffer'].apply(lambda b: compute_building_count(b, footprint_gdf))
buffer_area = 3.1416 * (buffer_radius ** 2)
gdf_points['Impervious_Ratio'] = gdf_points['Building_Area'] / buffer_area

# Merge these predictors into train_data (assuming the order of ground_df matches train_data).
train_data["Building_Area"] = gdf_points["Building_Area"].values
train_data["Building_Count"] = gdf_points["Building_Count"].values
train_data["Impervious_Ratio"] = gdf_points["Impervious_Ratio"].values
```

```python
# -------------------------------
# 7. Clean and Prepare Training Data
# -------------------------------
# Remove duplicate rows based on key spectral features (NDVI and LST).
columns_to_check = ['NDVI', 'LST']
for col in columns_to_check:
    train_data[col] = train_data[col].apply(lambda x: tuple(x) if isinstance(x, np.ndarray) and x.ndim > 0 else x)
train_data = train_data.drop_duplicates(subset=columns_to_check, keep='first')
train_data = train_data.reset_index(drop=True)
train_data = train_data.dropna()
print("Combined training data sample after cleaning:")
print(train_data.tail())

# Retain only the predictors and target.
# Predictors: NDVI, Albedo, LST, Bronx weather, Manhattan weather, Building_Area, Building_Count, Impervious_Ratio.
train_data = train_data[['NDVI', 'Albedo', 'LST', 
                         'Bronx_AirTemp', 'Bronx_WindSpeed', 'Bronx_SolarFlux',
                         'Manhattan_AirTemp', 'Manhattan_WindSpeed', 'Manhattan_SolarFlux',
                         'Building_Area', 'Building_Count', 'Impervious_Ratio', 
                         'UHI Index']]
```

```python
# -------------------------------
# 8. Splitting, Scaling, and Ensemble (Stacking) with Polynomial Features
# -------------------------------
# Separate predictors and target.
X = train_data.drop(columns=['UHI Index']).values
y = train_data['UHI Index'].values

# Add polynomial and interaction features (degree 2).
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
X_poly = poly.fit_transform(X)

# Split the data into training and testing sets.
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=123)

# Scale the polynomial features.
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train_poly)
X_test_scaled = sc.transform(X_test_poly)

# First, tune the XGBoost model.
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

xgb_model = XGBRegressor(random_state=42, objective='reg:squarederror')
random_search_xgb = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_grid,
    n_iter=50,
    cv=5,
    n_jobs=-1,
    scoring='r2',
    verbose=1,
    random_state=42
)
random_search_xgb.fit(X_train_scaled, y_train)
best_xgb = random_search_xgb.best_estimator_
print("Best XGBoost parameters:", random_search_xgb.best_params_)
print("Best cross-validated R2 score:", random_search_xgb.best_score_)

# Now, define a stacking ensemble combining the tuned XGB model, a RandomForest, and a Linear Regression.
base_estimators = [
    ('xgb', best_xgb),
    ('rf', RandomForestRegressor(n_estimators=200, random_state=42)),
    ('lr', LinearRegression())
]
stack_reg = StackingRegressor(estimators=base_estimators, final_estimator=LinearRegression(), cv=5, n_jobs=-1)

# Fit the stacking ensemble.
stack_reg.fit(X_train_scaled, y_train)

# Evaluate ensemble performance.
insample_preds = stack_reg.predict(X_train_scaled)
outsample_preds = stack_reg.predict(X_test_scaled)
print("Stacking Ensemble In-sample R2:", np.round(np.corrcoef(y_train, insample_preds)[0,1]**2, 4))
print("Stacking Ensemble Out-sample R2:", np.round(np.corrcoef(y_test, outsample_preds)[0,1]**2, 4))
```

```python
# ===============================
# SUBMISSION SECTION (Using Stacking Ensemble and Additional Predictors with Polynomial Features)
# ===============================
test_file = pd.read_csv('/Users/justin/Downloads/Submission_template.csv')
print("Submission template sample:")
print(test_file.head())

# Mapping Sentinel-2 satellite data for submission.
submission_sat = map_satellite_data('S2_sample.tiff', 'Submission_template.csv')
# submission_sat now contains Blue, Green, Red, NIR, NDVI, and Albedo.

# Mapping Landsat LST data for submission.
submission_lst = map_lst_data('landsat_lst.tiff', 'Submission_template.csv')

submission_data = pd.concat([submission_sat[['NDVI', 'Albedo']], submission_lst], axis=1)

# Add Bronx weather predictors.
bronx_weather_df = pd.read_csv("NY_Mesonet_Weather.csv")
bronx_summary = bronx_weather_df.agg({
    "Air Temp at Surface [degC]": "mean",
    "Avg Wind Speed [m/s]": "mean",
    "Solar Flux [W/m^2]": "mean"
}).to_dict()
submission_data["Bronx_AirTemp"]   = bronx_summary["Air Temp at Surface [degC]"]
submission_data["Bronx_WindSpeed"] = bronx_summary["Avg Wind Speed [m/s]"]
submission_data["Bronx_SolarFlux"] = bronx_summary["Solar Flux [W/m^2]"]

# Add Manhattan weather predictors.
manhattan_weather_df = pd.read_excel("NY_Mesonet_Weather.xlsx", sheet_name="Manhattan")
manhattan_summary = manhattan_weather_df.agg({
    "Air Temp at Surface [degC]": "mean",
    "Avg Wind Speed [m/s]": "mean",
    "Solar Flux [W/m^2]": "mean"
}).to_dict()
submission_data["Manhattan_AirTemp"]   = manhattan_summary["Air Temp at Surface [degC]"]
submission_data["Manhattan_WindSpeed"] = manhattan_summary["Avg Wind Speed [m/s]"]
submission_data["Manhattan_SolarFlux"] = manhattan_summary["Solar Flux [W/m^2]"]

# Compute building footprint and urban morphology predictors for submission.
gdf_sub = gpd.GeoDataFrame(test_file,
                           geometry=gpd.points_from_xy(test_file.Longitude, test_file.Latitude),
                           crs="EPSG:4326").to_crs(epsg=32618)
gdf_sub['buffer'] = gdf_sub.geometry.buffer(buffer_radius)

def compute_building_area(buffer_geom, footprints):
    intersections = footprints.intersection(buffer_geom)
    return intersections.area.sum()

def compute_building_count(buffer_geom, footprints):
    return footprints[footprints.intersects(buffer_geom)].shape[0]

gdf_sub['Building_Area'] = gdf_sub['buffer'].apply(lambda b: compute_building_area(b, footprint_gdf))
gdf_sub['Building_Count'] = gdf_sub['buffer'].apply(lambda b: compute_building_count(b, footprint_gdf))
gdf_sub['Impervious_Ratio'] = gdf_sub['Building_Area'] / (3.1416 * (buffer_radius ** 2))

submission_data["Building_Area"] = gdf_sub["Building_Area"].values
submission_data["Building_Count"] = gdf_sub["Building_Count"].values
submission_data["Impervious_Ratio"] = gdf_sub["Impervious_Ratio"].values

print("Submission predictors sample:")
print(submission_data.head())
print("Submission shape:", submission_data.shape)

# Apply the same polynomial transformation to the submission data.
submission_values = submission_data.values
submission_values_poly = poly.transform(submission_values)
transformed_submission = sc.transform(submission_values_poly)

# Predict the UHI Index.
final_predictions = stack_reg.predict(transformed_submission)
final_prediction_series = pd.Series(final_predictions)

submission_df = pd.DataFrame({
    'Longitude': test_file['Longitude'].values,
    'Latitude': test_file['Latitude'].values,
    'UHI Index': final_prediction_series.values
})
print("Final submission DataFrame:")
print(submission_df.head())
submission_df.to_csv("submission.csv", index=False)
```
